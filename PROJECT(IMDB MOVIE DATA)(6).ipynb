```python

def replace_outliers(dataframe,feature):   
    Q1=dataframe[feature].quantile(0.25)
    Q3=dataframe[feature].quantile(0.75)
    IQR=Q3-Q1
    upper_bound=Q3+1.5*IQR  
    lower_bound=Q1-1.5*IQR
    
                                                                                # capping-replacing all higher values with upper bound (75% percentile)
    dataframe[feature]=np.where(dataframe[feature]>upper_bound,upper_bound,dataframe[feature])
                                                                                # flooring -replace all lower values with lower bounf (25% percentile)
    dataframe[feature]=np.where(dataframe[feature]<lower_bound,lower_bound,dataframe[feature])
```

```python
replace_outliers(df,"Runtime (Minutes)")
```

1) Average Rating by Main Genre       (Shows which genres score best.)

```python
avg_rating_genre = df.groupby('Genre')['Rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(
    x=avg_rating_genre.values,
    y=avg_rating_genre.index,
    hue=avg_rating_genre.values,   # add hue
    palette='coolwarm',
    dodge=False,                   # avoid grouping effect
    legend=False                   # remove legend if not needed
)
plt.title('Average Rating by Main Genre')
plt.xlabel('Average Rating')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()


```

```python
##The chart makes it easy to compare which movies earned the most, with the longest bars representing the highest revenues.
```

2. Top ten movies by IMDB RATING              (To highlight the most critically acclaimed movies)

```python
top10_rating = df.sort_values(by='Rating', ascending=False).head(10)

plt.figure(figsize=(8, 5))
sns.barplot(x='Title', y='Rating', data=top10_rating, palette='coolwarm', hue='Title', dodge=False, legend=False)
plt.title("Top 10 Movies by IMDb Rating")
plt.ylabel("IMDb Rating")
plt.xlabel("Movie Title")
plt.xticks(rotation=45, ha='right')  # Rotate x labels for clarity
plt.tight_layout()
plt.show()
```

```python
##This chart shows the top 10 movies in our dataset ranked by IMDb rating.
##I sorted the dataset in descending order of rating and selected the top 10 movies.
##The horizontal bars represent the rating score for each movie, making it easy to compare their ratings.
##This chart highlights the most critically acclaimed movies, which may or may not match the top revenue list — showing that popularity and quality ##don’t always go hand in hand.
```

3.Top 10 Directors with Most Movies             (To see which directors are most frequent)

```python
top_directors = df['Director'].value_counts().head(10)

sns.barplot(x=top_directors.values, y=top_directors.index)
plt.title("Top 10 Directors with Most Movies")
plt.xlabel("Number of Movies")
plt.ylabel("Director")
plt.tight_layout()
plt.show()
```

```python
## This chart shows the top 10 directors with the most movies in our dataset.
## I counted how many times each director appears and selected the top 10.
## The horizontal bars represent the number of movies each director has directed in our dataset, giving us an idea of the most prolific filmmakers."
```

4.Top 10 most frequent actors 

```python
from collections import Counter
```

```python
actor_list = df['Actors'].str.split(', ').sum()
top_actors = Counter(actor_list).most_common(10)
actors_df = pd.DataFrame(top_actors, columns=['Actor', 'Count'])
```

```python

plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Actor', data=actors_df)
plt.title('Top 10 Most Frequent Actors')
plt.xlabel('Count')
plt.ylabel('Actor')
plt.tight_layout()
plt.show()
```

```python
## First, I extracted all actor names from the dataset by splitting the 'Actors' column into individual names and combining them into one list.
## Then, I used Python’s Counter function to count how many times each actor appears and selected the top 10 most frequent actors.
## Finally, I visualized this information in a bar chart. This chart clearly shows which actors appear the most in our dataset, which can indicate their popularity or frequent casting in the industry."
```

5.Number of Movies Released Per Year (To analyze the trend of movie releases over time)

```python
year_counts = df['Year'].value_counts().sort_index()

plt.figure(figsize=(8, 8))
plt.pie(year_counts, labels=year_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Movies Released Per Year")
plt.show()
```

```python
## This pie chart shows the percentage of movies released each year in our dataset.
## Each slice represents a different year, with its size showing how much it contributes to the total number of movies.
## This visualization helps compare movie production across years in percentage terms rather than exact counts.
```
